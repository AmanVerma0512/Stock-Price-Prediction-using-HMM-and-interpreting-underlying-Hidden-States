```python
df['Difference'] = df['Close'].diff(-1)
df['Difference Mod'] = df['Difference'].abs()
```

```python
criteria = [df['Difference'].between(-2500, -35), df['Difference'].between(-35, 35), df['Difference'].between(35, 3935)]
values = ['Down', 'Stagnant', 'Up']

df['Movement'] = np.select(criteria, values, 'Stagnant')
```
